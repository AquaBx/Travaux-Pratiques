```python
t = np.linspace(-2, 2, 1000)

def triangle(liste):
    return [ 1 - np.abs(x)  if np.abs(x) < 1 else 0 for x in liste]

def box(liste):
    return [ 1  if np.abs(x) < 0.5 else 0 for x in liste]

def calculmain(liste):
    r = []
    for x in liste:
        if -1.5 < x <= -0.5:
            r.append((2*x+3)**2/8)
        elif -0.5 < x <= 0.5:
            r.append(3/4-(x**2))
        elif 0.5 < x <= 1.5:
            r.append((2*x-3)**2/8)
        else : 
            r.append(0)
    return r

convo = np.convolve(triangle(t),box(t), mode = 'same')/len(t)*4

            
plt.plot(t,triangle(t))
plt.plot(t,box(t))
plt.show()
plt.plot(t,convo)
plt.plot(t,calculmain(t))
plt.show()
```

```python
def unitaire(liste):
    return [ 1  if x >= 0 else 0 for x in liste]

def expon(liste):
    return [ (np.e ** (-2*x) if x >= 0 else 0 ) for x in liste]

def calculmain(liste):
    r = []
    for x in liste:
        if 0 < x:
            r.append((1-np.e**(-2*x))/2)
        else : 
            r.append(0)
    return r

convo = np.convolve(unitaire(t),expon(t), mode = 'same')/len(t)*4

plt.plot(t,unitaire(t))
plt.plot(t,expon(t))
plt.show()
plt.plot(t,convo)
plt.plot(t,calculmain(t))
plt.show()
```

on trouve la même chose sur les deux car les courbes se superposent
